```python
import pandas as pd
import json


with open('/content/sample_data/IN_category_id[1].json', 'r') as f:
    data = json.load(f)


category_dict = {}
for item in data['items']:
    category_dict[int(item['id'])] = item['snippet']['title']

df = pd.read_csv('/content/sample_data/INvideos[1].csv')


df['category_name'] = df['category_id'].map(category_dict)


print("Successfully Mapped! here the sample:")
print(df[['title', 'category_name']].head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12,8))

sns.countplot(data=df, y='category_name', order=df['category_name'].value_counts().index, palette='viridis')

plt.title('India YouTube Trending Categories Analysis')
plt.xlabel('Number of Trending Videos')
plt.ylabel('Video Category')

plt.show()
```

```python
correlation = df['views'].corr(df['likes'])
print(f"Views and Likes madhya correlation: {correlation}")

plt.figure(figsize=(10,6))
sns.scatterplot(x=df['likes'], y=df['views'])
plt.title('Likes vs Views Correlation')
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
import numpy as np

X = df[['likes']]
y = df['views']

model = LinearRegression()
model.fit(X, y)

test_likes = np.array([[100000]])
predicted_views = model.predict(test_likes)

print(f"Prediction:If one video have 1 lakhs likes, then it may get  {int(predicted_views[0])} views!")
```
